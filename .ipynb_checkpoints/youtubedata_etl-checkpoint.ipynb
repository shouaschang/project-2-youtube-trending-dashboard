```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
import psycopg2
```

### Store CSV into DataFrame

```python
csv_file = "./data/USvideos_edit2.csv"
video_data_df = pd.read_csv(csv_file)
video_data_df.head(30)
```

```python
df_by_id = video_data_df.drop_duplicates(['video_id'], keep='last')
df_by_id
```

### Connect to local database

```python
rds_connection_string = "postgres:bright@localhost:5432/Youtube_data"
engine = create_engine(f'postgresql://{rds_connection_string}')
# connection= engine.connect()
# result= connection.execute()
inspector = inspect(engine)
inspector.get_columns("Youtube_videos")
```

### Use pandas to load csv converted DataFrame into database

```python
video_data_df.to_sql(name='Youtube_videos', con=engine, if_exists='append', index=False)
```
